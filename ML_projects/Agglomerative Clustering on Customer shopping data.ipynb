```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
```

```python
customer_data = pd.read_csv('shopping_data.csv')
print(customer_data.head())
print(customer_data.shape)
print(type(customer_data))
```

```python
# We use Dendrograms from scipy.cluster.hierarchy to find out the optimal number of clusters
# The longest continuous horizontal line that can be drawn on the dendrogram gives the optimal cluster number.  
data = customer_data.iloc[:,3:5].values
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
# Applying Agglomerative Clustering on the customer_data
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c = cluster.labels_, cmap='rainbow')
```
